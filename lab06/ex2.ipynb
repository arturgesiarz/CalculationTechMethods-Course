```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import roots_legendre
```

### Funkcja do calkowania

```python
def function_to_integrate(x):
    return 4 / (1 + x ** 2)
```

### Implemetacja metody Gaussa-Legendre'a

```python
def gauss_legendre_integration(n):
    nodes, weights = roots_legendre(n)

    x = 0.5 * (nodes + 1)
    w = 0.5 * weights

    integral_value = np.sum(w * function_to_integrate(x))

    return integral_value
```

### Obliczenie wartości dokładnej całki

```python
exact_value = np.pi
```

### Lista przechowująca wartości bezwzględnych błędów względnych

```python
errors = []
```

### Zakres wartości liczby węzłów n

```python
n_values = range(1, 100)
```

### Obliczenie wartości całki i błędów dla różnych wartości n

```python
for n in n_values:
    integral_value = gauss_legendre_integration(n)
    error = np.abs((exact_value - integral_value) / exact_value)
    errors.append(error)
```

### Narysowanie wykresu

```python
plt.figure(figsize=(10,6))
plt.plot(np.array(n_values) + 1, errors, label='Metoda Gaussa-Legendre’a',marker='o')
plt.xscale('log')
plt.yscale('log')
plt.title('Błąd względny w zależności od liczby ewaluacji funkcji')
plt.xlabel('Liczba ewaluacji funkcji')
plt.ylabel('Bezwzględny błąd względny')
plt.legend()
plt.grid(True)
plt.show()
```
